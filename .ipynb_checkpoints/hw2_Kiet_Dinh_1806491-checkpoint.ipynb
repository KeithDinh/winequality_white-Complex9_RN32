```python

def to_class(df_list):
    new_col =[]
    for i in df_list:
        if int(i) >= 8 and i<= 10:
            new_col.append('A')
        elif int(i) == 7:
            new_col.append('B')
        elif int(i) == 6:
            new_col.append('C')
        elif int(i) == 5:
            new_col.append('D')
        elif int(i) <= 4:
            new_col.append('E')
    return new_col
df['class'] = to_class(df['quality'])
df
```

```python
#Reference: Anh Nguyen, Dat Nguyen, George Nav
def entropy(a,b):
    partition = []
    for i in range(len(a)):
        while a[i]>=len(partition):
            partition.append([])
        partition[a[i]].append(b[i])
    a=partition
    b=list(set(b))
    
    len_a_flat=0
    for each in a:
        len_a_flat = len_a_flat+len(each)
    entr=0
    for c in range(1, len(a)):
        c_entr = 0
        p= np.zeros([len(b)])
        if len(a[c]) != 0:
            for o in a[c]:
                for i in range(len(b)):
                    if o == b[i]:
                        p[i] = p[i]+1
            for i in range(len(p)):
                if p[i] != 0:
                    p[i]=p[i]/len(a[c])
                    c_entr = c_entr+(-p[i]*math.log(p[i],2))
        entr = entr +(len(a[c]))/(len_a_flat-len(a[0]))*c_entr
    return (entr, len(a[0])/len_a_flat)
```

```python
#Reference: Anh Nguyen, Dat Nguyen, George Nav

def ordinal_variation(a,b):
    ord_var = 0
    phi = {'A':4,
           'B':3,
           'C':2,
           'D':1,
           'E':0}
    num_groups = max(a)
    partition = []
    for i in range(len(a)):
        while a[i]>=len(partition):
            partition.append([]);
        partition[a[i]].append(b[i])
    a = partition
    b = list(set(b))
       
    len_a_flat = 0
    for c in a:
        len_a_flat = len_a_flat+len(c)
    for c in range(1,len(a)):
        ov = 0
        if len(a[c])!=1 and len(a[c])!=0:
            for i in range(len(a[c])):
                for j in range(len(a[c])):
                    if i!=j:
                        ov = ov+abs(phi[a[c][i]]-phi[a[c][j]])
            ov = ov/(math.pow(len(a[c]), 2)-len(a[c]))
            ord_var = ord_var+(len(a[c])/(len_a_flat-len(a[0])))*ov
    return ord_var
```

```python
print(ordinal_variation([0,1,1,1,1,2,2,3],['A','A','A','E','E','D','D','C']))
```

```python
#Reference: Anh Nguyen, Dat Nguyen, George Nav

def variance(a,b):
    partition = []
    for i in range(len(a)):
        while a[i]>= len(partition):
            partition.append([])
        partition[a[i]].append(b[i])
    a=partition
    b=list(set(b))
    len_a_flat= 0
    for c in a:
        len_a_flat=len_a_flat+len(c)
    var=0
    for c in range(1,len(a)):
        if len(a[c])>1:
            v =0
            mean = np.mean(a[c])
            for i in a[c]:
                v= v+ (math.pow(i-mean,2))
            v=v/(len(a[c])-1)
            var=var+(len(a[c]))/(len_a_flat-len(a[0]))*v
    return (var, len(a[0])/len_a_flat)
```

```python
print(variance([0,1,1,1,1,2,2,3],[8,8,8,4,4,5,5,6]))
```

```python
from scipy.spatial.distance import pdist, squareform
def mdist(d):
    l = pd.DataFrame()
    cols = list(d.columns)
    for z in cols:
        col_zscore = str(z) + '_zscore'
        l[col_zscore]=(d[z] - d[z].mean())/d[z].std(ddof=0)
    return squareform(pdist(l.values, metric='cityblock'))
    
newdf = pd.read_csv("winequality-white.csv")
newdf = mdist(newdf.drop(['quality'], axis=1))
print("The manhattan distance between data in the first 11th attributes")
newdf
```

### The code below generates a loop with 1000 iterations. Each loop will create a list of 11 random numbers from range 0 to 9223372036854775807 (max system). Then multiply each number to each column in the 11 attributes. Then it will compute the kmeans and find predicted 12th attributes. It computes the variance and compare to the previous variance. If the new variance is lower, the old variance will be updated to new, and save the new list

```python
var_temp = 10
w_vector =[]
for m in range (1000):
    if(m == 100):
        print(m)
    x_wwq_8 = df.drop(['quality','class'],axis =1)
    weight = [random.randint(0,9223372036854775807) for _ in range(11)]
    for i,j in zip(cols_to_norm, weight):
        x_wwq_8[i] = x_wwq_8[i]*j
    kmeans = KMeans(n_clusters=8,random_state=123).fit(x_wwq_8)
    y_wwq_8 = kmeans.predict(x_wwq_8)
    var8,per8 = variance(list(y_wwq_8),list(df['quality']))
    if(var8 < var_temp):
        var_temp = var8
        w_vector = weight
        print(var_temp)
```

```python
print("Variance: " + str(var_temp))
entr_test,entr_per = entropy(list(y_wwq_8), to_class(df['quality']))
print("Entropy: "+ str(entr_test))
ordinal = entropy(list(y_wwq_8), to_class(df['quality']))
print("ordinal_variation: "+ str(ordinal))
```
