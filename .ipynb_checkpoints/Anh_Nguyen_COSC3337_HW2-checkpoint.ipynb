```python
#Question 1

def entropy(a,b):
    num_groups = max(a)
    partition = []
    for i in range(len(a)):
        while a[i]>= len(partition):
            partition.append([])
        partition[a[i]].append(b[i])
    a = partition
    b = list(set(b))
    
    len_a_flat = 0
    for c in a:
        len_a_flat = len_a_flat+len(c)
    entr=0
    for c in range(1, len(a)):
        c_entr=0
        p = np.zeros([len(b)])
        if len(a[c])!=0:
            for o in a[c]:
                for i in range(len(b)):
                    if o==b[i]:
                        p[i] = p[i]+1
            for i in range (len(p)):
                if p[i]!=0:
                    p[i] = p[i]/len(a[c])
                    c_entr = c_entr+(-p[i]*math.log(p[i], 2))
        entr = entr+(len(a[c])/(len_a_flat-len(a[0])))*c_entr
    return(entr, len(a[0])/len_a_flat)
```

```python
#Testing Case 1
a = (0,1,1,1,1,2,2,3)
b= ('A','A','A','E','E','D','D','C')
entropy(a,b)
```

```python
#Question 2

def ordinal_variation(a,b):
    ord_var = 0
    phi = {'A':4,
           'B':3,
           'C':2,
           'D':1,
           'E':0}
    num_groups = max(a)
    partition = []
    for i in range(len(a)):
        while a[i]>=len(partition):
            partition.append([]);
        partition[a[i]].append(b[i])
    a = partition
    b = list(set(b))
        
    len_a_flat = 0
    for c in a:
        len_a_flat = len_a_flat+len(c)
    for c in range(1,len(a)):
        ov = 0
        if len(a[c])!=1 and len(a[c])!=0:
            for i in range(len(a[c])):
                for j in range(len(a[c])):
                    if i!=j:
                        ov = ov+abs(phi[a[c][i]]-phi[a[c][j]])
            ov = ov/(math.pow(len(a[c]), 2)-len(a[c]))
            ord_var = ord_var+(len(a[c])/(len_a_flat-len(a[0])))*ov
    return ord_var
```

```python
#Testing Case 1
a = (0,1,1,1,1,2,2,3)
b= ('A','A','A','E','E','D','D','C')
ordinal_variation(a,b)
```

```python
#Testing Case 2
a=(1,1,1,0,0,2,2,2)
b= ('A','A','A','E','E','D','D','C')
ordinal_variation(a,b)
```

```python
#Question 3

def variance(a, b):
    num_groups = max(a)
    partition = []
    for i in range(len(a)):
        while a[i]>=len(partition):
            partition.append([])
        partition[a[i]].append(b[i])
    a = partition
    b = list(set(b))
    
    len_a_flat = 0
    for c in a:
        len_a_flat = len_a_flat+len(c)
    var = 0
    for c in range(1, len(a)):
        if len(a[c])>1:
            v = 0
            mean = np.mean(a[c])
            for i in a[c]:
                v = v+(math.pow(i-mean, 2))
            v = v/(len(a[c])-1)
            var = var+(len(a[c])/(len_a_flat-len(a[0])))*v
    return (var, len(a[0])/len_a_flat)
```

```python
#Testing Case 1
a = (0,1,1,1,1,2,2,3)
c = (8,8,8,4,4,5,5, 6)
variance(a,c)
```

```python
from scipy.spatial.distance import pdist, squareform
def mdist(d):
    newlist = pd.DataFrame()
    cols = list(d.columns)
    for z in cols:
        zscore = str(z) + '_zscore'
        newlist[zscore]=(d[z] - d[z].mean())/d[z].std(ddof=0)
    return squareform(pdist(newlist.values, metric='cityblock'))
    
```

```python
# Import key libraries
import seaborn as sns

# Fit KMeans and calculate SSE for each *k*
sse = {}
for k in range(0, 5):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(df6)
    sse[k] = kmeans.inertia_ # sum of squared distances to closest cluster cente
```

```python
# Plot SSE for each *k*
plt.title('The Elbow Method')
plt.xlabel('k'); plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
# K-Means determine k
from scipy.spatial.distance import cdist
wwq = pd.read_csv("winequality-white.csv")
norm_cols = ['fixed.acidity','volatile.acidity','citric.acid','residual.sugar','chlorides','free.sulfur.dioxide','total.sulfur.dioxide','density','pH' ,'sulphates' ,'alcohol']
wwq[norm_cols] = StandardScaler().fit_transform(wwq[norm_cols])
wwq.drop(columns=['quality'])

distortions = []
K = range(1,11)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(wwq)
    kmeanModel.fit(wwq)
    distortions.append(np.sum(np.min(cdist(wwq, kmeanModel.cluster_centers_, 'cityblock'), axis=1))/wwq.shape[0])
    sse[k] = kmeanModel.inertia_ # sum of squared distances to closest cluster cente
sse
```

```python
# Plot the elbow
plt.plot(K,distortions,'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method Showing the Optimal K')
plt.show()
```

```python
test_y = np.linspace(0,2*np.pi,100)
test_x = np.cos(df7)
# Determine which parts are close to zero
close_value = np.isclose(test_x, 3874, atol=1e-1)

chunks = []
prev_close = False
for x, y, close in zip(test_x, test_y, close_value):
    if not close:
        if prev_close:
            chunks.append(chunk)
    if close:
        if not prev_close:
            chunk=[[y], [x]]
        chunk[0].append(y)
        chunk[1].append(x)
    prev_close = close
chunks
```
